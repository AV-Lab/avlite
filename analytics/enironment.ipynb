```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import shapely
from shapely.geometry import Polygon
import icecream as ic
```

```python
agent_corners = np.array([[0,0],[0,1],[1,2],[1,0]])
ego_corners = np.array([[-.5,-.5], [-.5,.5], [.5,.5], [.5,-.5]])
agent = plt.Polygon(agent_corners, closed=True, edgecolor='r', facecolor='azure')
ego = plt.Polygon(ego_corners)

agent_poly = Polygon(agent_corners) 
ego_poly = Polygon(ego_corners)

collision = ego_poly.intersects(agent_poly)
print(collision)

fig, ax1 = plt.subplots()
ax1.set_xlim(-1,2)
ax1.set_ylim(-1,2)

ax1.add_patch(agent)
ax1.add_patch(ego)



                
```
