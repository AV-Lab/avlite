```python
import matplotlib.pyplot as plt
import json 
```

```python
track_boundaries_path = "resource/yasmarina.track.json"

with open(track_boundaries_path, 'r') as f:
    track_boundaries = json.load(f)

reference_file = "resource/yas_marina_race_line_mue_0_5_4_m_margin.json"
with open(reference_file, 'r') as f:
    reference1 = json.load(f)
reference_file = "resource/yas_marina_real_race_line_mue_0_5_3_m_margin.json"
with open(reference_file, 'r') as f:
    reference2 = json.load(f)

reference_file = "/home/mkhonji/workspaces/A2RL_Integration/config/tracks/yas_south_322_safe_test.json"
with open(reference_file, 'r') as f:
    reference3 = json.load(f)
```

```python
import numpy as np

left_boundary = track_boundaries["LeftBound"]
right_boundary = track_boundaries["RightBound"]
reference_line1 = reference1["ReferenceLine"]
reference_line3 = reference3["ReferenceLine"]
reference_point = reference3["ReferencePoint"]


plt.figure(figsize=(20, 20))  

left_x, left_y = zip(*[(point[0], point[1]) for point in left_boundary])
plt.plot(left_x, left_y, color='orange', label='Left Boundary') 


right_x, right_y = zip(*[(point[0], point[1]) for point in right_boundary])
plt.plot(right_x, right_y, color='tan', label='Right Boundary') 

reference_x, reference_y = zip(*[(point[0], point[1]) for point in reference_line1])
plt.plot(reference_x, reference_y, color='blue', label='Reference 1') 

reference_x, reference_y = zip(*[(point[0], point[1]) for point in reference_line3])
plt.plot(reference_x, reference_y, color='green', label='South') 

plt.plot(reference_point[0], reference_point[1], 'ro', label='Reference Point')
# zoom = 500
# plt.xlim(24.46992202098782 - zoom, 24.46992202098782 + zoom)
# plt.ylim(54.60522506805341 - zoom, 54.60522506805341 + zoom)

plt.legend() 
plt.show()




print(reference1.keys())
```
